```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity. 
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()`For a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)). 




```python
# compare the average income based on ethnicity
print("The average income of each ethnicity is\n", df.groupby('Ethnicity')['Income'].mean())

# on average, do married or single people have a higher balance?
single = df[df.Married == 'No'].loc[:,'Income'].mean()
married = df[df.Married == 'Yes'].loc[:,'Income'].mean()
print("\nThe average income of a single person is:", single)
print("The average income of a married person is:", married)
print("Therefore, a married person has a higher balance.")

# what is the highest and lowest income in our dataset?
print("\nThe highest income in our dataset is:", df.loc[:,'Income'].max())
print("The lowest income in our dataset is:", df.loc[:,'Income'].min())

# how many cards do we have recorded in our dataset?
print("\nWe have", df.loc[:,'Cards'].sum(), "cards in our dataset.")

# how many females do we have information for vs. how many for males?
female = df[df.Gender == 'Female'].loc[:,'Gender'].count()
male = df[df.Gender == 'Male'].loc[:,'Gender'].count()
print("\nWe have information for", female, "females, as opposed to", male, "males.")
```
